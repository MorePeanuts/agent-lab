```python
from langchain.messages import HumanMessage
from langchain.chat_models import init_chat_model

model = init_chat_model('deepseek-chat', max_tokens=20)
prompts = [
    'Why do parrots have colorful feathers?',
    'How do airplanes fly?',
    'What is quantum computing?',
]
```

```python
responses1 = model.batch(prompts, config={'max_concurrency': 3})  # type: ignore
```

```python
type(responses1), type(responses1[0])
```

```python
for msg in responses1:
    msg.pretty_print()
```

```python
for response in model.batch_as_completed(prompts, config={'max_concurrency': 3}):
    print(response[0])
    response[1].pretty_print()
```

#### Agent batch output

```python
from langchain.agents import create_agent

agent = create_agent(model)
```

```python
responses2 = agent.batch([{'messages': HumanMessage(prompt)} for prompt in prompts])
for response in responses2:
    response['messages'][-1].pretty_print()
```
